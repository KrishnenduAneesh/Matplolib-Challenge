## Summary Statistics

## Quartiles, Outliers and Boxplots

```python

# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
specified_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Filter for the specified treatment regimens
filtered_data = combined_data[combined_data['Drug Regimen'].isin(specified_regimens)]

# Find the last (greatest) timepoint for each mouse
max_timepoints = filtered_data.groupby(['Mouse ID'])['Timepoint'].max().reset_index()

# Merge to get the tumor volume at the last timepoint for the specified regimens
final_tumor_volume = pd.merge(max_timepoints, filtered_data, on=['Mouse ID', 'Timepoint'], how='inner')

# Display the final tumor volume of each mouse in the specified treatment regimens
print(final_tumor_volume[['Drug Regimen', 'Mouse ID', 'Tumor Volume (mm3)']])
```

```python
#Put treatments into a list for for loop (and later for plot labels)
# List of treatments for the loop and plot labels
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create an empty list to store tumor volume data
tumor_vol_data = []

#Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
# Locate the rows which contain mice on each drug and get the tumor volumes

for treatment in treatments:
    # Filter the data for each treatment
    treatment_data = combined_data[combined_data['Drug Regimen'] == treatment]
    
    # Get the tumor volumes for this treatment
    tumor_volumes = treatment_data['Tumor Volume (mm3)']
    
    # Add subset data to the list
    tumor_vol_data.append(tumor_volumes)
    
    #Calculate the IQR and determine potential outliers
    quartiles = tumor_volumes.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)
    
    # Determine outliers using upper and lower bounds
    outliers = tumor_volumes[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    
    # Print results
    #print(f"-----{treatment} Treatment-----")
    print(f"{treatment} Potential outliers for : {outliers}\n")
    print(f"IQR for {treatment}: {iqr}")
    
    #print(f"Values below {lower_bound} could be outliers.")
    #print(f"Values above {upper_bound} could be outliers.")
   
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.figure(figsize=(8, 6))
plt.boxplot(tumor_vol_data, labels=treatments)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_id = 'y793'  

# Filter data for the selected mouse treated with Capomulin
mouse_data = combined_data[(combined_data['Mouse ID'] == mouse_id) & (combined_data['Drug Regimen'] == 'Capomulin')]

# Plot tumor volume vs. time point for the selected mouse
plt.figure(figsize=(8, 6))
plt.plot(mouse_data['Timepoint'], mouse_data['Tumor Volume (mm3)'], marker='o', color='blue')
plt.title(f'Capomulin treatment of mouse {mouse_id}')
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data = combined_data[combined_data['Drug Regimen'] == 'Capomulin']

# Calculate the average tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merge average tumor volume with the original Capomulin data to include weight
capomulin_with_avg_volume = capomulin_data.merge(average_tumor_volume, on='Mouse ID', suffixes=['', '_avg'])

# Remove duplicate Mouse ID entries and keep only unique records
unique_capomulin_data = capomulin_with_avg_volume.drop_duplicates(subset='Mouse ID')

# Create a scatter plot of mouse weight vs. the average observed tumor volume
plt.figure(figsize=(8, 6))
plt.scatter(unique_capomulin_data['Weight (g)'], unique_capomulin_data['Tumor Volume (mm3)_avg'], color='blue')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

import seaborn as sns
from scipy import stats
correlation_coefficient = st.pearsonr(unique_capomulin_data['Weight (g)'], unique_capomulin_data['Tumor Volume (mm3)_avg'])[0]

# Create a linear regression model and plot
print(f"The correlation between mouse weight and the average tumor volume is  {correlation_coefficient}")
plt.figure(figsize=(8, 6))
sns.regplot(x='Weight (g)', y='Tumor Volume (mm3)_avg', data=unique_capomulin_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()

```
